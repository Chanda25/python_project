```python
import pandas as pd
df= pd.read_csv('orders.csv', na_values=['Not Available' ,'unknown'] )
# treat NA and unknown values as a null we use na_values while reading file
```

```python
df.head(5)
```

```python
df.isnull().sum() #check null
df.dropna() #drop null records
df.describe() # check the data overall(descriptive statistics)
```

```python
df['Ship Mode'].value_counts()
```

```python
#rename the all column names
df.columns=df.columns.str.lower() # replace the column name with the cap to lower case
df.columns = df.columns.str.replace(' ' , '_')  # replace the space with '_'
df.head(2)
```

```python
#derive the discount column from the cost_price and discount_percent 
df['total_discount'] = df['cost_price'] * df['discount_percent']/100
df.head(2)
```

```python
#derive the sale price from colm list_price and discount
df['sale_price'] = df['cost_price'] - df['total_discount']
df.head(2)
```

```python
#derive profit ==>saleing price - cost price
df['profit'] = df['cost_price'] -df['sale_price']
df.head(2)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'] , format = "%Y-%m-%d")
#change the datatype of date table from fomat to date
```

```python
#drop the columns from table but stored in different datafreame
new_df=df.drop(columns=['list_price' , 'cost_price' , 'discount_percent'])
new_df.head(2)
```

# 1. Find top 10 Higest revenue generating products

```python
df['revenue']=df['sale_price'] * df['quantity']  # calculated revenue
total_sales = df.groupby('product_id')['revenue'].sum().reset_index() #grouped by product_id and total of revenue
top_revenue= total_sales.nlargest(5 , 'revenue') # get top most revenue
print(top_revenue)  # printed top revenue
```

# 2. Find top 5 highest seiling product in each region

```python
seiling_product= df.groupby(['region' , 'product_id'])['sale_price'].sum().reset_index() 
top_seiling = seiling_product.sort_values(['region','sale_price'], ascending=[True , False]).groupby('region').head(5)
print(top_seiling)
```

# 3.find month-over-month growth comparison for 2022 , 2023 sales  eg : jan 2022 vs jan 2023

```python
#fetch the year and month from order date
df['year']= df['order_date'].dt.year 
df['month'] = df['order_date'].dt.month
df.head(2)
```

```python
#group by monthly_sales by year and month
monthly_sales = df.groupby(['year' , 'month'])['sale_price'].sum().reset_index()
```

```python
#create pivot table for comparision
pivot_sales= monthly_sales.pivot(index='month' , columns = 'year' , values = 'sale_price').fillna(0)
#pivot_sales['growth'] = ((pivot_sales[2023] - pivot_sales[2022]) / pivot_sales[2022]).replace([float('inf') , -float('inf')] ,0)*100 
```

```python
print(pivot_sales[[2022,2023]])
```

# 4. for each category which month has the highest sales


```python
#groupby data on year,month,catergory and sum of sales_price
sales = df.groupby(['year' , 'month' , 'category'])['sale_price'].sum().reset_index()
highest_sales= sales.loc[sales.groupby('category')['sale_price'].idxmax()]
print(highest_sales)
```
