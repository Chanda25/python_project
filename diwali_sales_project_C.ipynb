```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating countplot==>> for title we have to use set() function and for getting the count we have to use for loop and stores
#values in containers..

xs = sns.countplot(x= 'Gender'  , data = df)
xs.set(title= 'Count of gender')
for bars in xs.containers:
    xs.bar_label(bars)
plt.show()
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount' ,ascending=False)
sales_gen
```

```python
sx=sns.barplot(x= 'Gender' ,y='Amount', data = sales_gen)
plt.show()
```

```python
#From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
#AGE
ag=sns.countplot(x='Age Group' ,hue='Gender', data = df)

for ags in ag.containers:
    ag.bar_label(ags)

plt.show()
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'],as_index =False)['Amount'].sum().sort_values(by ='Amount' , ascending=True)
sales_age
```

```python
bar=sns.barplot(x='Age Group' , y='Amount' , data=sales_age)
bar.set_title("Age group v/s Amount")
plt.show()
```

```python
#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#State
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sales_state
```

```python
sns.barplot(x='State' ,y = 'Orders' ,data = sales_state)
bar.set_title("State wise orders")
sns.set(rc ={'figure.figsize':(15,5)})
plt.show()
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sales_state
```

```python
sns.barplot(x='State' ,y = 'Amount' ,data = sales_state)
bar.set_title("State wise Amount")
sns.set(rc ={'figure.figsize':(15,5)})
plt.show()
```

```python
#From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh,
#Maharashtra and Karnataka respectively
```
